```python
df = pd.read_csv( index_col = 0,  parse_dates = True, infer_datetime_format = True, 
                         filepath_or_buffer = './data/2018-01-01__2019-01-01__NConservatory__allMerged.csv')
df.index = pd.to_datetime(df.index, utc=True).tz_convert('America/Los_Angeles')
```

```python
externConditionLabelEncoder = LabelEncoder();
externConditionLabelEncoder.fit(['clear', 'cloudy', 'partlyCloudy', 'rain', 'showers', 'thunderstorms', 'ice', 'flurries', 'snow']);
```

```python
df.dropna(axis=0, how='all', inplace=True)
df.dropna(axis=1, how='all', inplace=True)
```

```python
timestamps = pd.to_datetime(df.index, infer_datetime_format=True, utc=True)
df["hour"] = timestamps.hour
df["dayofweek"] = timestamps.dayofweek
df["weekofyear"] = timestamps.weekofyear
```

```python
df.head(25)
```

```python
%matplotlib inline
df.hist(df.columns, figsize=(20,60), layout=(len(df.columns),1))
```

```python
for col in ["dayofweek","weekofyear","hour","externCondition_1"]:
    one_hot = pd.get_dummies(df[col], prefix=col, dummy_na=False)
    one_hot.drop([col for col, val in one_hot.sum().iteritems() if val < 10], axis=1, inplace=True)
    df.drop(col, inplace=True, axis=1, errors='ignore')
    df = df.join(one_hot.loc[:,one_hot.sum(axis=0)>1], lsuffix='', rsuffix='_'+col)
    print(one_hot.sum())
df.shape
```

```python
df = df.T.drop_duplicates().T
df.shape
```

```python
y_col = "co2_1"
y = np.log1p(df[y_col].values)
X = df.drop([y_col], inplace=False, axis=1, errors='ignore').values
t = df.index.values
```

```python
num_rows = len(y)
split_size = 0.10
split_idx = int(split_size*num_rows)
X_train = X[:split_idx]
X_test = X[split_idx:]
y_train = y[:split_idx]
y_test = y[split_idx:]
t_train = t[:split_idx]
t_test = t[split_idx:]
```

```python
from sklearn.linear_model import RidgeCV
clf = RidgeCV(alphas=[1e-3, 1e-2, 1e-1, 1]).fit(X_train, y_train)
clf.score(X_train, y_train) 
```

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
y_test = np.expm1(y_test)
y_pred = np.expm1(y_pred)
mean_squared_error(y_test, y_pred)
```

```python
anomaly = [100 if (y_pred[i]-yt)/y_pred[i] > 0.25 else 1 for i,yt in enumerate(y_test)]
colors = ['green' if a==1 else 'red' for a in anomaly]
```

```python
%matplotlib inline
fig, ax = plt.subplots()
fig.set_figheight(15)
fig.set_figwidth(100)

ax.plot(t_test, y_pred, label="Expected")
ax.scatter(t_test, y_test, s=anomaly, color=colors, label="Actual")

ax.legend(loc='best')

plt.xlim(t_test[0], t_test[-1])

plt.show()
```
